```python
import re
import pandas as pd
from bs4 import BeautifulSoup
import urllib
```

```python
url = "http://rotoguru1.com/cgi-bin/byday.pl?game=dk&month=7&day=1&year=2019"
```

```python
page = urllib.request.urlopen(url)
soup = BeautifulSoup(page,'html.parser')
```

```python
header = soup.find_all('u')
head = []
for var in header:
    cols = var.find_all('b')
    cols = [ele.text.strip() for ele in cols]
    head.append(cols)
head = head[:7]
head[0] = ['Pos']
head
```

```python
player = soup.find_all('a', href=re.compile("http://rotoguru1.com/cgi-bin/player"))
play = []
for var in player:
    col = var.text
    play.append(col)
play
```

```python
# Points and Team

t = soup.find_all('tr')
test = []
for ele in t:
    te = ele.find_all('td', {'align':'center'})
    te = [elem.text.strip() for elem in te]
    test.append(te)
    
test

str_list = list(filter(None, test)) # fastest
str_list
```

```python
# Salary

s = soup.find_all('tr')
sal = []
for ele in s:
    sa = ele.find_all('td', {'align':'right'})
    sa = [elem.text.strip() for elem in sa]
    sal.append(sa)

salary = list(filter(None, sal)) # fastest
salary = salary[2:]
salary
```
