# Step 1 - Open Browswer and go to page

```python
from selenium.webdriver import Firefox
from selenium.webdriver.firefox.options import Options
# opts = Options()
# opts.set_headless()
# assert opts.headless  # Operating in headless mode
browser = Firefox(executable_path = "C:/Users/finnxm/Documents/geckodriver-v0.24.0-win64/geckodriver.exe")
browser.get('https://www.fangraphs.com/dailyprojections.aspx?pos=all&stats=bat&type=sabersim')
```

# Step 2 - click on 1000

```python
browser.find_element_by_xpath("//*[@id='DFSBoard1_dg1_ctl00_ctl02_ctl00_PageSizeComboBox_Arrow']").click()
```

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
```

```python
WebDriverWait(browser, 20).until(EC.visibility_of_element_located((By.XPATH, "//*[@id='DFSBoard1_dg1_ctl00_ctl02_ctl00_PageSizeComboBox_DropDown']/div/ul/li[7]"))).click()
```

# Step 3 - Get Column Names

```python
col = browser.find_element(By.XPATH, '//*[@id="DFSBoard1_dg1_ctl00"]/thead/tr')
column_names = col.text.split()
column_names
```

# Step 4 - Get Data

```python
dat = []

table = browser.find_element(By.XPATH, '//*[@id="DFSBoard1_dg1_ctl00"]/tbody') 
for rows in table.find_elements(By.TAG_NAME, 'tr'):
    ele = [td.text for td in rows.find_elements(By.TAG_NAME, 'td')]
    dat.append(ele)

dat
```

# Step 5 - Place in Pandas

```python
import pandas as pd
sabersim_df = pd.DataFrame(data=dat, columns=column_names, dtype=float)
```
