```python
import pandas as pd
from random import randint
from time import sleep
import datetime
import io
import requests
import numpy as np
```

```python
def rotogrinders_projections():
    header = ['Player', 'Salary', 'Team', 'Pos', 'Opp', 'Max', 'Min', 'proj_pts']
    bat_url = "https://rotogrinders.com/projected-stats/mlb-hitter.csv?site=draftkings"
    bat_s=requests.get(bat_url).content
    rotogrinders_hitters_df=pd.read_csv(io.StringIO(bat_s.decode('utf-8')),
                                       header=None,
                                       names=header)
    #rotogrinders_hitters_df = pd.read_csv("https://rotogrinders.com/projected-stats/mlb-hitter.csv?site=draftkings",
     #                                 header=None,
     #                                 names=header)
    
    sleep(randint(1,3))
    
    pitch_url = "https://rotogrinders.com/projected-stats/mlb-pitcher.csv?site=draftkings"
    pitch_s = requests.get(pitch_url).content
    rotogrinders_pitchers_df = pd.read_csv(io.StringIO(pitch_s.decode('utf-8')),
                                      header=None,
                                      names=header)
    
    frames = [rotogrinders_hitters_df, rotogrinders_pitchers_df]
    
    rotogrinders_proj_df = pd.concat(frames, ignore_index=False)
    rotogrinders_proj_df['Date'] = datetime.datetime.today().strftime('%Y-%m-%d') 
    
    del frames
    
    return(rotogrinders_proj_df)
```

```python
df = rotogrinders_projections()
```

```python

ax = sns.scatterplot(x='Salary', y='proj_pts', hue='Player', data=pitchers)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
```

# Clustering

```python
from sklearn.cluster import KMeans
```

```python
pitcher_value = pd.DataFrame(pitchers['proj_pts'] / pitchers['Salary'], columns=['Value'])

pitcher_tiers = KMeans(n_clusters=3).fit(np.array(pitcher_value["Value"]).reshape(-1,1))
pitcher_series = pd.DataFrame(pitcher_tiers.labels_, columns = ['Tier'])
pitchers  = pd.concat([pitchers, pitcher_series], axis=1)
pitchers
```

```python
ax = sns.scatterplot(x='proj_pts', y='Salary', hue='Tier', data=pitchers)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
```
