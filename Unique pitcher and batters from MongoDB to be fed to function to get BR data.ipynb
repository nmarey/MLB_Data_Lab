```python
import pymongo 
import pandas as pd
```

```python
client = pymongo.MongoClient('localhost', 27017)
db = client['MLB']
db_collection = db['statcast']
```

```python
pitcher_df = pd.DataFrame(list(db_collection.distinct('pitcher')), columns=["pitcher"])
pitcher_df
```

```python
batter_df = pd.DataFrame(list(db_collection.distinct('batter')), columns=['batter'])
batter_df
```

```python
client = pymongo.MongoClient('localhost', 27017)
db = client['MLB']
db_collection_player_map = db['player_name_and_ids']
```

```python
query_dict = {'mlb_id': batter_df['batter'][0].item() }
query_dict
```

```python
fields_return_dict = {"_id": 0, "mlb_id": 1, "bref_id": 1}
```

```python
df = pd.DataFrame(list(db_collection_player_map.find(query_dict,fields_return_dict))) #convert entire collection to pandas DataFrame
df 
```

```python
fields_return_dict = {"_id": 0, "mlb_id": 1, "bref_id": 1}
data = []
for k in batter_df['batter']:
    query_dict = {'mlb_id': k}
    result = pd.DataFrame(list(db_collection_player_map.find(query_dict,fields_return_dict)))
    data.append(result)
    
data = pd.concat(data, axis=0, ignore_index=True)

data.info()
```

```python
fields_return_dict = {"_id": 0, "mlb_id": 1, "bref_id": 1, "throws": 1, "espn_pos": 1, "debut": 1}
pitcher_query = []
for k in pitcher_df['pitcher']:
    query_dict = {'mlb_id': k}
    result = pd.DataFrame(list(db_collection_player_map.find(query_dict,fields_return_dict)))
    pitcher_query.append(result)
    
pitcher_query = pd.concat(pitcher_query, axis=0, ignore_index=True)

pitcher_query.info()
```
