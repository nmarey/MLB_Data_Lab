# Pitcher Repertoire

## Pitches Thrown

```python
client = pymongo.MongoClient('localhost', 27017)
db = client['MLB']
db_collection=db['statcast']

query_dict = {'pitcher' : 608331}

fields_return_dict = { "_id": 0, "pitch_type": 1, "launch_angle": 1, "pitch_name":1, "bb_type":1, "description" : 1, 'p_throws' : 1, 'stand' : 1, 'events' : 1}

repertoire_df = pd.DataFrame(list(db_collection.find(query_dict,fields_return_dict)))

client.close()

repertoire_df
```

```python
pitches_df = repertoire_df.pitch_type.value_counts(normalize=True)
pitches_df = pitches_df.reset_index()
pitches_df
pitches_df = pitches_df.rename({'index': 'p_type', 'pitch_type' : 'pitch_percentage'}, axis='columns')
pitches_df = pitches_df.set_index('p_type')
pitches_df = pitches_df.transpose()
pitches_df = pitches_df.reset_index()
pitches_df = pitches_df.drop(columns='index')
pitches_df
```

## Pitches Outcome

```python
from sklearn import preprocessing
```

```python
pitch_result_df = repertoire_df[pd.notna(repertoire_df['events'])]
pitch_result_df = pitch_result_df[['events', 'pitch_type']]
pitch_result_df = pitch_result_df.groupby(['pitch_type', 'events'], as_index = False).size().reset_index(name="count")
pitch_result_df['p_type_event'] = pitch_result_df['pitch_type']+"_"+pitch_result_df['events']
pitch_result_df = pitch_result_df.drop(columns=['pitch_type', 'events'])
pitch_result_df

#x = pitch_result_df['count'].values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(np.array(pitch_result_df['count']).reshape(-1, 1))
pitch_result_df['outcomes'] = x_scaled
pitch_result_df
# pitch_result_df value_counts(normalize=True)
```
