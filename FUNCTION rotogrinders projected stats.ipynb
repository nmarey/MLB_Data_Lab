```python
def rotogrinders_projections():
    import pandas as pd
    from random import randint
    from time import sleep
    import datetime
    import io
    import requests
    
    header = ['Player', 'Salary', 'Team', 'Pos', 'Opp', 'Max', 'Min', 'proj_pts']
    bat_url = "https://rotogrinders.com/projected-stats/mlb-hitter.csv?site=draftkings"
    bat_s=requests.get(bat_url).content
    rotogrinders_hitters_df=pd.read_csv(io.StringIO(bat_s.decode('utf-8')),
                                       header=None,
                                       names=header)
    #rotogrinders_hitters_df = pd.read_csv("https://rotogrinders.com/projected-stats/mlb-hitter.csv?site=draftkings",
     #                                 header=None,
     #                                 names=header)
    
    sleep(randint(1,3))
    
    pitch_url = "https://rotogrinders.com/projected-stats/mlb-pitcher.csv?site=draftkings"
    pitch_s = requests.get(pitch_url).content
    rotogrinders_pitchers_df = pd.read_csv(io.StringIO(pitch_s.decode('utf-8')),
                                      header=None,
                                      names=header)
    
    frames = [rotogrinders_hitters_df, rotogrinders_pitchers_df]
    
    rotogrinders_proj_df = pd.concat(frames, ignore_index=False)
    rotogrinders_proj_df['Date'] = datetime.datetime.today().strftime('%Y-%m-%d') 
    
    del frames
    
    return(rotogrinders_proj_df)
```

```python
df = rotogrinders_projections()
```

```python
client = MongoClient('localhost', 27017)
db = client['MLB']
collection = db['rotogrinders_projections']
   
count_before = collection.count_documents({})
    
result = collection.insert_many(df.to_dict(orient='records'))
    
count_after = collection.count_documents({})
    
client.close()
```

```python
print(["Count before: {}".format(count_before), "Count after: {}".format(count_after)])
```

```python
import pandas as pd
from pymongo import MongoClient
import MLB_CONFIG

def move_to_db(df_name, collection_name, db_name = MLB_CONFIG.DATABASE_CONFIG['db_name']):
    client = MongoClient('localhost', 27017)
    db = client[db_name]
    collection = db[collection_name]
    
    count_before = collection.count_documents({})
    
    result = collection.insert_many(df_name.to_dict(orient='records'))
    
    count_after = collection.count_documents({})
    
    client.close()
    
    output = ["Count before: {}".format(count_before), "Count after: {}".format(count_after)]
    
    return(output)
```
